```python
import random 
q = 97
n = 4
noise = 2
A = [[random.randint(0, q-1) for _ in range(n)] for _ in range(n)]
s = [random.randint(0, q-1) for _ in range(n)]
e = [random.randint(-noise, noise) for _ in range(n)]

b = []
for i in range(n):
    value = sum(A[i][j] * s[j] for j in range(n)) + e[i]
    b.append(value % q)

print("Public key A:", A)
print("Public key b:", b)
print("Secret key s:", s)

message = 1

r = [random.randint(0, 1) for _ in range(n)]
e1 = [random.randint(-noise, noise) for _ in range(n)]
e2 = random.randint(-noise, noise)

u = [(sum(A[j][i] * r[j] for j in range(n)) + e1[i]) % q for i in range(n)]
v = (sum(b[i] * r[i] for i in range(n)) + e2 + (q//2)*message) % q

print("\nCiphertext:")
print("u =", u)
print("v =", v)

value = (v - sum(s[i] * u[i] for i in range(n))) % q

if q//4 < value < 3*q//4:
    print("\nDecrypted message: 1")
else:
    print("\nDecrypted message: 0")
```
